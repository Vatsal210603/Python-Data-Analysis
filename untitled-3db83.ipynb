# Rename the column Distance to Landmark to Distance

```python
data=data.rename(columns={'Distance_to _Landmark':'Distance'})
```

```python
data
```

# Determine the location type

```python
def typelocation(Distance):
    if Distance <2:
       return "Prime_Location"
    elif Distance<5:
       return "Good_Location"
    elif Distance<10:
        return " Average_Location"
    else:
       return " Poor_Location"
     
```

```python
data['Location_Type']=data['Distance'].apply(typelocation)
```

```python
data.head()
```

# Show the relation Location type and Nearest Landmark for hotels

```python
sns.countplot(x='Location_Type',hue='Nearest_Landmark',data=data)
plt.show()
```

```python
plt.figure(figsize=(16,5))
sns.distplot(data['Price'],bins=10)
plt.show()
```

# Show a combined relation between Price and Tax and mention the potential outliers
